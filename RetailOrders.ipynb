```python
import pandas as pd
df= pd.read_csv('orders.csv')
df
```

```python
#Checking for any Null Values
df.isnull().values.any()
```

```python
#Sum of the Null Values
df.isnull().sum()
```

```python
#Filling the Null Values
df_new = pd.DataFrame(df)
df_filled = df.fillna({'Ship Mode':'unknown'})
print(df_filled)
```

```python
#Again checking for Null Values
df_filled.isnull().sum()
```

```python
df_filled.head()
```

```python
#Renaming the columns to Lower case
df_filled.columns = df_filled.columns.str.lower()
```

```python
df_filled.columns
```

```python
df=pd.DataFrame(df_filled)
```

```python
df.rename(columns={'order id':'order_id','order date':'order_date','ship mode':'ship_mode','postal code':'postal_code',
                          'sub category':'sub_category','product id':'product_id','cost price':'cost_price','list price':'list_price',
                          'discount percent':'discount_percent'})
```

```python
df.columns
```

```python
#Trimming Spaces in text
df.columns = df.columns.str.replace(' ','_')
df.head(5)
```

```python
#Calculating Discounts -- New column discount
df['discount'] = df['list_price']*df['discount_percent']*.01
```

```python
#New Column: SALE PRICE
df['sale_price'] = df['list_price']-df['discount']
```

```python
#New Column: PROFIT
df['profit'] = df['sale_price']-df['cost_price']
```
